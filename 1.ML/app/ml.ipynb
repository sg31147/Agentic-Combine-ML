```python
import pandas as pd
import requests
merged_table = pd.read_csv("test_dataset.csv")
```

```python
url='http://localhost:8081/predict'
```

```python
import json

for idx, row in merged_table.iterrows():
    if not pd.isna(row["ml"] or row["ml_description"]):
        continue

    payload = {
        "Task": "Ranking",
        "Precisionk": 15,
        "id": [27218502],
        "text": [row['text']],
        "target": [[]],
        "split": "test"
    }

    try:
        response = requests.post(url, json=payload, headers={'Content-Type': 'application/json'})
        response.raise_for_status()
        json_response = response.json()

        if "data" in json_response:
            first_key = next(iter(json_response["data"]))
            result = json_response["data"][first_key].get("result", {})
        else:
            result = {}

    except Exception as e:
        print("Error:", e)
        result = {"error": str(e)}

    merged_table.at[idx, "ml_description"] = json.dumps(result)
    merged_table.at[idx, "ml"] = json.dumps(list(result.keys()))
```

```python
merged_table.to_csv("ml.csv", index=False)
merged_table.to_csv("../../2.LM/claude/Claude.csv", index=False)
merged_table.to_csv("../../3.Agent_Combine_ML/claude/Claude.csv", index=False)
```
